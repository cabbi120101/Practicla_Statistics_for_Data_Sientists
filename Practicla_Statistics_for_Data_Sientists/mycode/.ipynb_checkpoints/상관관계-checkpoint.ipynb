## 상관관계
- 상관계수(피어슨 상관계수) : 수치적 변수들 간에 어떤 관계가 있는지를 나타내기 위해 사용되는 측정량(-1~1 사이값)
- 변수들이 선형적인 관계를 가지지 않는 경우, 상관관계는 유용한 측정 지표가 아니다.

```python
sp500_sym = pd.read_csv(SP500_SECTORS_CSV)
sp500_px = pd.read_csv(SP500_DATA_CSV, index_col=0)
```

```python
# Table 1-7
# Determine telecommunications symbols
telecomSymbols = sp500_sym[sp500_sym['sector'] == 'telecommunications_services']['symbol']

# Filter data for dates July 2012 through June 2015
telecom = sp500_px.loc[sp500_px.index >= '2012-07-01', telecomSymbols]
telecom.corr()
print(telecom)
```

```python
fig, ax = plt.subplots(figsize=(5, 4))
ax = sns.heatmap(etfs.corr(), vmin=-1, vmax=1, 
                 cmap=sns.diverging_palette(20, 220, as_cmap=True),
                 ax=ax)

plt.tight_layout()
plt.show()
```

```python
from matplotlib.collections import EllipseCollection
from matplotlib.colors import Normalize

# 타원모양의 방향은 두 변수가 양의 관계를 갖는지, 음의 관계를 갖는지 나타낸다.
# 타원의 색깔과 너비는 상관관계의 강도를 의미, 얇고 진할수록 더 강한 관계성을 나타낸다.

def plot_corr_ellipses(data, figsize=None, **kwargs):
    ''' https://stackoverflow.com/a/34558488 '''
    M = np.array(data)
    if not M.ndim == 2:
        raise ValueError('data must be a 2D array')
    fig, ax = plt.subplots(1, 1, figsize=figsize, subplot_kw={'aspect':'equal'})
    ax.set_xlim(-0.5, M.shape[1] - 0.5)
    ax.set_ylim(-0.5, M.shape[0] - 0.5)
    ax.invert_yaxis()

    # xy locations of each ellipse center
    xy = np.indices(M.shape)[::-1].reshape(2, -1).T

    # set the relative sizes of the major/minor axes according to the strength of
    # the positive/negative correlation
    w = np.ones_like(M).ravel() + 0.01
    h = 1 - np.abs(M).ravel() - 0.01
    a = 45 * np.sign(M).ravel()

    ec = EllipseCollection(widths=w, heights=h, angles=a, units='x', offsets=xy,
                           norm=Normalize(vmin=-1, vmax=1),
                           transOffset=ax.transData, array=M.ravel(), **kwargs)
    ax.add_collection(ec)

    # if data is a DataFrame, use the row/column names as tick labels
    if isinstance(data, pd.DataFrame):
        ax.set_xticks(np.arange(M.shape[1]))
        ax.set_xticklabels(data.columns, rotation=90)
        ax.set_yticks(np.arange(M.shape[0]))
        ax.set_yticklabels(data.index)

    return ec, ax

m, ax = plot_corr_ellipses(etfs.corr(), figsize=(5, 4), cmap='bwr_r')
cb = fig.colorbar(m, ax=ax)
cb.set_label('Correlation coefficient')

plt.tight_layout()
plt.show()
```

### 다변량 분석
- 분할표 : 두가지 이상의 범주형 변수의 빈도수 기록한 표
- 육각형 구간 : 두 변수를 육각형 모양의 구간으로 나눈 그림
- 등고 도표 : 두 변수의 밀도를 등고선으로 표시한 도표
- 바이올린 도표 : 밀도 추정을 함께 보여줌

```python
kc_tax = pd.read_csv(KC_TAX_CSV)
kc_tax0 = kc_tax.loc[(kc_tax.TaxAssessedValue < 750000) & 
                     (kc_tax.SqFtTotLiving > 100) &
                     (kc_tax.SqFtTotLiving < 3500), :]
print(kc_tax0.shape)
```

### 육각형 구간과 등고선 (수치형 변수 대 수치형 변수 시각화)
- 데이터량이 너무 커지면 산점도의 점들이 너무 밀집되어 보기 어렵다.

```python
# hexbin 사용
ax = kc_tax0.plot.hexbin(x='SqFtTotLiving', y='TaxAssessedValue',
                         gridsize=30, sharex=False, figsize=(5, 4))
ax.set_xlabel('Finished Square Feet')
ax.set_ylabel('Tax Assessed Value')

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(4, 4))
sns.kdeplot(data=kc_tax0.sample(10000), x='SqFtTotLiving', y='TaxAssessedValue', ax=ax)
ax.set_xlabel('Finished Square Feet')
ax.set_ylabel('Tax Assessed Value')

plt.tight_layout()
plt.show()
```

### 범주형 변수 대 범주형 변수
- 분할 표는 범주별 빈도수를 기록한 표다.
- 분할표를 통해 각 열의 빈도나 전체 백분율을 알 수 있다.

```python
lc_loans = pd.read_csv(LC_LOANS_CSV)
```

```python
# Table 1-8(1)
crosstab = lc_loans.pivot_table(index='grade', columns='status', 
                                aggfunc=lambda x: len(x), margins=True)
print(crosstab)
```

```python
# Table 1-8(2)
df = crosstab.copy().loc['A':'G',:]
# 행합계로 행을 나눈다.
df.loc[:,'Charged Off':'Late'] = df.loc[:,'Charged Off':'Late'].div(df['All'], axis=0)
# 
df['All'] = df['All'] / sum(df['All'])
perc_crosstab = df
print(perc_crosstab)
```

### 다변수 시각화 하기
- 조건화라는 개념을 통해 두 변수 비교용 도표(산점도, 육각형 구간, 상자 구간)를 더 확장하여 활용할 수 있다.

```python
zip_codes = [98188, 98105, 98108, 98126]
kc_tax_zip = kc_tax0.loc[kc_tax0.ZipCode.isin(zip_codes),:]
kc_tax_zip

def hexbin(x, y, color, **kwargs):
    cmap = sns.light_palette(color, as_cmap=True)
    plt.hexbin(x, y, gridsize=25, cmap=cmap, **kwargs)

g = sns.FacetGrid(kc_tax_zip, col='ZipCode', col_wrap=2)

# map으로 다른 우편번호에 대한 원래 데이터 집합의 subset에 대해 hexbin 함수 호출
# extent는 x,y축의 한계를 정의한다.
g.map(hexbin, 'SqFtTotLiving', 'TaxAssessedValue', 
      extent=[0, 3500, 0, 700000])
g.set_axis_labels('Finished Square Feet', 'Tax Assessed Value')
g.set_titles('Zip code {col_name:.0f}')

plt.tight_layout()
plt.show()
```
